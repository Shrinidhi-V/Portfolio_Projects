```python
dy = df.groupby('Year').sum()
dy = dy.drop(['State'], axis = 1)
dy
```

```python
dy.loc[:,'Total_Crimes'] = dy.sum(numeric_only=True, axis=1)
dy
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,6))
sns.barplot(data = dy, x = 'Year', y = 'Total_Crimes', color = 'Red')
plt.title('Year-wise Total Crime against Women from 2001 - 2021')
plt.show()
```

```python
# Plotting Rape incidents over years for each state
for state in df['State'].unique():
    subset = df[df['State'] == state]
    plt.plot(subset['Year'], subset['Rape'], marker='o', label=state)

plt.title('Rape Incidents Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.show()
```

```python
df.set_index('Year')[['Rape', 'K&A', 'DD', 'AoW', 'AoM', 'DV', 'WT']].plot.area()
plt.figure(figsize= (12,8))
plt.title('Cumulative Crime Incidents Over Years')
plt.ylabel('Number of Incidents')
plt.xlabel('Year')
plt.show()
```

```python
plt.scatter(df['DV'], df['DD'])
plt.title('Domestic Violence vs Dowry Deaths')
plt.xlabel('Domestic Violence Incidents')
plt.ylabel('Dowry Deaths')
plt.show()
```

```python
sns.set_style('whitegrid')
df_2021 = df[df['Year'] == 2021].set_index('State')
df_2021[['Rape', 'K&A', 'DD', 'AoW', 'AoM', 'DV', 'WT']].plot(kind='bar')
plt.title('Crime Incidents by State in 2021')
plt.ylabel('Number of Incidents')
plt.xlabel('State')
plt.show()
```

```python
df_2021_total = df_2021[['Rape', 'K&A', 'DD', 'AoW', 'AoM', 'DV', 'WT']].sum()
df_2021_total.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Crime Types in 2021')
plt.ylabel('')
plt.show()
```

```python
df_2020 = df[df['Year'] == 2020].set_index('State')
df_2020_total = df_2020[['Rape', 'K&A', 'DD', 'AoW', 'AoM', 'DV', 'WT']].sum()
df_2020_total.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Crime Types in 2020')
plt.ylabel('')
plt.show()
```
